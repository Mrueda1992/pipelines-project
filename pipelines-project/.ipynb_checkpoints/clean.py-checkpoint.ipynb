```python
import pandas as pd

```

```python
def get_cols(df):
    cols = df[['Year','Player','Pos','Age','Tm','PER','3PAr','3P', '3PA', '3P%',
       '2P', '2PA', '2P%', 'eFG%','PTS']]
    
```

```python
read_csv = pd.read_csv('nba-stats.csv')
data = pd.DataFrame(read_csv)
data.head()
data.shape
data.columns
#data.isnull().sum()
```

```python
#solo queremos estadísticas de tiro para medir a los mejores tiradores de triples
df = pd.DataFrame()
df = data[['Year','Player','Pos','Age','Tm','PER','3PAr','3P', '3PA', '3P%',
       '2P', '2PA', '2P%', 'eFG%','PTS']]
#limpiamos nulos
df = df[df['3P'].notnull()]
df = df[df['3PAr'].notnull()]
df = df[df['3P%'].notnull()]

df.head()
    
```

```python
df.rename(index=str, columns= {"Pos":"Position","Tm":"Team"}).head(1)
```

```python
#filtramos por años superiores a 1980
over_2009 = df['Year']>=2009
nba= df[over_2009]
nba.head(10)
```

```python
#máximo anotador de triples desde 2009

three_pts = nba.groupby('Player')['3P'].sum().sort_values(ascending=False)
three_shoots = nba.groupby('Player')['3PA'].sum().sort_values(ascending=False)
threes = three_pts.to_frame().join(three_shoots.to_frame())
threes['3Rate'] = round(three_pts/three_shoots,3)
threes.head()

```
